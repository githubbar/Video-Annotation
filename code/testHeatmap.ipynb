```python
from scipy.ndimage.filters import gaussian_filter
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set_theme()
import time

fig, axs = plt.subplots(ncols=2)
data = np.random.rand(240, 320)
sns.heatmap(data, square=True, vmin=-0, vmax=1, cmap ="coolwarm" , ax=axs[0])
t0 = time.time()
data_smooth = gaussian_filter(data, sigma=5)
t1 = time.time()
print(f'min={data_smooth.min()} max={data_smooth.max()}')
sns.heatmap(data_smooth, square=True, vmin=data_smooth.min(), vmax=data_smooth.max(), cmap ="coolwarm" , cbar=True, ax=axs[1]) 
print(f'total time to smooth={t1-t0} sec')
```
